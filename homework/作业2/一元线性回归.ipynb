```python
import numpy as np   #数值计算
import matplotlib.pyplot as plt        #画图

# 数据
x = np.array([1, 2, 3, 4, 5])
y = np.array([40, 60, 65, 85, 90])

# 算均值
x_mean = np.mean(x)
y_mean = np.mean(y)

# 算斜率 b
b = np.sum((x - x_mean) * (y - y_mean)) / np.sum((x - x_mean) ** 2)   #公式

# 算截距 a
a = y_mean - b * x_mean

# 打印结果
print(f"b = {b}")
print(f"a = {a}")
print(f"y = {b}x + {a}")

# 预测新的值
x_new = 6
y_new = b * x_new + a
print(f"预测: x={x_new}时, y={y_new}")

# 画图      图片出现中文会有乱码
plt.scatter(x, y, color='red', s=100, label='date')   #散点图
plt.plot(x, b * x + a, color='blue', label='line')    #折线图/曲线
plt.scatter(x_new, y_new, color='green', marker='*', s=200, label='predict')   #预测图
plt.xlabel('learntime')         #x轴的标签
plt.ylabel('score')        #y轴的标签
plt.title(f'y = {b}x + {a}')        #图表标题
plt.legend()           #显示图例放左上角
plt.grid(True)             #背景网格线
plt.show()         #渲染元素   显示图片
```
